```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the accuracy data from acc.log
acc_file_path = 'acc.log'
accuracies = []

# Read the accuracy values from the log file
with open(acc_file_path, 'r') as file:
    for line in file:
        accuracies.append(float(line.strip().replace('%', '')))

# Create a DataFrame to store the accuracy data
epochs = list(range(1, len(accuracies) + 1))
acc_data = pd.DataFrame({
    'Epoch': epochs,
    'Accuracy': accuracies
})

# Set the style and font
plt.style.use('seaborn-darkgrid')
plt.rcParams['font.family'] = 'Georgia'
plt.rcParams['font.size'] = 14

# Plot the training accuracy curve
plt.figure(figsize=(12, 8))
plt.plot(acc_data['Epoch'], acc_data['Accuracy'], marker='o', linestyle='-', color='b', label='Accuracy')
plt.title('Training Accuracy Curve', fontsize=18)
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Accuracy (%)', fontsize=16)
plt.legend(fontsize=14)
plt.grid(True)

# Annotate the training time on the plot
training_time = "Training time 3:27:22"
plt.text(0.5, 0.5, training_time, transform=plt.gca().transAxes, fontsize=14, verticalalignment='center', horizontalalignment='center', bbox=dict(facecolor='white', alpha=0.5))

plt.show()
```
